Para esta parte utilizaremos el conjunto de datos Countries, el cual contiene almacenado un conjunto de paises donde cada uno es descrito por una serie de datos numéricos.

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
rand = 911 #Numero para los parametros aleatorios

df = pd.read_csv("countries.csv")
df = df.drop(['Region'], axis=1)
df = df.dropna()
df.head()
```

```python
df.describe()
```

Seleccionamos todos los atributos que tengan un indice de correlación mayor que 0.6.

```python
#Filtro
correl=df.corr()
correl[abs(correl['GDP (USD per capita)']) > 0.6]['GDP (USD per capita)']
```

```python
df = df[['Country','Infant mortality (per 1000 births)','Phones (per 1000)','Birthrate','Agriculture',
         'GDP (USD per capita)']]

coun = df[['Infant mortality (per 1000 births)','Phones (per 1000)','Birthrate','Agriculture']].values
gdp = df['GDP (USD per capita)'].values
```

```python
pd.plotting.scatter_matrix(df, figsize = (15, 15), diagonal = 'kde', color = 'red')
```

Seleccionamos las 4 variables que pasaron el filtro y combrobamos como en la tabla como las curvas son todas semejantes entre si, siendo la variable Phones la que mas se asemeja a GDP.

Aunque no sea estricramente necesario, estandarizamos los datos para la mejora del tiempo de entrenamiento y evitar los mínimos locales en la optimización del error.

```python
# Transformamos las variables de entrada
scaler_x = StandardScaler()
scaler_x.fit(coun)
coun_scaled=scaler_x.transform(coun)

# Y también transformamos la variable de salida
scaler_y = StandardScaler()
scaler_y.fit(gdp.reshape(-1, 1)) # Como es una sola variable es necesario ponerlo en modo columna
gdp_scaled=scaler_y.transform(gdp.reshape(-1, 1))
```

```python
regul_param_range = 10.0 ** -np.arange(1, 7) 

cv_scores_mean=[]
cv_scores_std=[]

for regul_param in regul_param_range:
    # aumentamos el parámetro max_iter hasta que converja
    mlp=MLPRegressor(hidden_layer_sizes=(200, ), activation='relu', solver='adam', alpha=regul_param, batch_size=100, 
             learning_rate='constant', learning_rate_init=0.001, max_iter=600, random_state=rand)
    
    # El error es neg_mean_squared_error el MSE en negativo!! Más pequeño peor.
    scores = cross_val_score(mlp, coun_scaled, gdp_scaled.ravel(), cv=5, scoring='neg_mean_squared_error')
    
    cv_scores_mean.append(scores.mean())
    cv_scores_std.append(scores.std())
```

```python
# Dibujamos la línea de la exactitud media en las partes de test
plt.plot(np.log10(regul_param_range), cv_scores_mean, color="g", label="Test")

# Dibujamos la banda de la desviación típica
lower_limit=np.array(cv_scores_mean)-np.array(cv_scores_std)
upper_limit=np.array(cv_scores_mean)+np.array(cv_scores_std)
plt.fill_between(np.log10(regul_param_range), lower_limit, upper_limit, color="#DDDDDD")

# Creamos el gráfico
plt.title("Curva de aprendizaje")
plt.xlabel("Regularización 10^{-X}"), plt.ylabel("negative MSE"), plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

```python
regul_param_range = 10.0 ** -np.arange(1, 7) 

cv_scores_mean=[]
cv_scores_std=[]

for regul_param in regul_param_range:
    # aumentamos el parámetro max_iter hasta que converja
    mlp=MLPRegressor(hidden_layer_sizes=(50, ), activation='relu', solver='adam', alpha=regul_param, batch_size=100, 
             learning_rate='constant', learning_rate_init=0.001, max_iter=500, random_state=rand)
    
    # El error es neg_mean_squared_error el MSE en negativo!! Más pequeño peor.
    scores = cross_val_score(mlp, coun_scaled, gdp_scaled.ravel(), cv=5, scoring='neg_mean_squared_error')
    
    cv_scores_mean.append(scores.mean())
    cv_scores_std.append(scores.std())
```

```python
# Dibujamos la línea de la exactitud media en las partes de test
plt.plot(np.log10(regul_param_range), cv_scores_mean, color="g", label="Test")

# Dibujamos la banda de la desviación típica
lower_limit=np.array(cv_scores_mean)-np.array(cv_scores_std)
upper_limit=np.array(cv_scores_mean)+np.array(cv_scores_std)
plt.fill_between(np.log10(regul_param_range), lower_limit, upper_limit, color="#DDDDDD")

# Creamos el gráfico
plt.title("Curva de aprendizaje")
plt.xlabel("Regularización 10^{-X}"), plt.ylabel("negative MSE"), plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

El perceptrón multicapa de 200 neuronas tiene un comportamiento en general mejor, tanto en la media como en la desviacíon típica. Al tener una cantidad 4 veces más de neuronas ayuda en tener resultados más finos.

```python
regul_param=10**-4

mlp=MLPRegressor(hidden_layer_sizes=(200, ), activation='relu', solver='adam', alpha=regul_param, batch_size=100, 
             learning_rate='constant', learning_rate_init=0.001, max_iter=600, random_state=rand)
    

predicts_scaled = cross_val_predict(mlp, coun_scaled, gdp_scaled.ravel(), cv=5)


# escalamos las predicciones para que estén en la escala original
predicts=scaler_y.inverse_transform(predicts_scaled)
```

```python
# Dibujamos el histograma de los errores
fig1 = plt.figure()
ax1 = plt.axes()
ax1.hist(np.abs(gdp - predicts))  # np.abs(boston.target - predicts) es el error en cada ejemplo

plt.title('Distribución del error en valor absoluto')
plt.show()
```

```python
fig2 = plt.figure()
ax2 = plt.axes()

# dibujamos los errores: la distancia a la recta indica el nivel de error
ax2.scatter(gdp, predicts) 
# Pintamos la referencia de la predicción perfecta
ax2.plot(gdp, gdp, color='k') 

plt.title('Y Vs Y_pred')
plt.show()
```

```python
new_df = df.copy()
new_df['Predicted GDP'] = np.around(predicts)
```
