```python
from sklearn import datasets
import pandas as pd
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, export_graphviz, plot_tree
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_validate
import numpy as np
```

```python
cancer = datasets.load_breast_cancer()
cancer['feature_names']
```

```python
df = pd.DataFrame(data=cancer['data'], columns=cancer['feature_names']) 
df['type'] = cancer['target']
dic_target = { 0 : 'malignant', 1 : 'benign'}
df['type'] = df['type'].map(dic_target)
df.head()
```

```python
df['type'].value_counts().plot.pie(figsize=(7, 7), autopct = '%.1f%%')
```

```python
sns.set(style="ticks")
sns.pairplot(df, hue='type', vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area','mean smoothness', 'mean compactness', 'mean concavity','mean concave points', 'mean symmetry', 'mean fractal dimension'])
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(cancer.data, cancer.target)
```

```python
plt.figure(figsize=(25,25))
plot_tree(clf, filled=True, feature_names=cancer.feature_names, class_names=cancer.target_names, rounded=True)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, test_size=0.3, random_state=911)

# Creamos y entrenamos el árbol de decisión
clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

# Calculamos la precisión del modelo de entrenamiento y de test
train_accuracy = clf.score(X_train, y_train)
test_accuracy = clf.score(X_test, y_test)
train_accuracy, test_accuracy
```

```python
train_accuracy = []
test_accuracy = []

max_depths = range(1, 6)
for md in max_depths: 
    # Entrenar y validar
    clf = DecisionTreeClassifier(criterion="entropy", max_depth=md, min_samples_split=2)
    scores = cross_validate(clf, cancer.data, cancer.target, scoring='accuracy', cv=10, return_train_score=True)
    
    train_accuracy.append(np.mean(scores['train_score']))
    test_accuracy.append(np.mean(scores['test_score']))
```

```python
# Draw lines
plt.plot(max_depths, train_accuracy, color="r",  label="Training")
plt.plot(max_depths, test_accuracy, color="g", label="Test")

# Create plot
plt.title("Curva de aprendizaje")
plt.xlabel("Parametro"), plt.ylabel("Accuracy Score"), plt.legend(loc="best")
plt.tight_layout()
plt.show()   
```

```python
clf = DecisionTreeClassifier(criterion="entropy",  # por defecto Gini pero podemos cambiar a entropía
                             max_depth=4,          # profundidad máxima del árbol
                             min_samples_split=5,  # mínimo de muestras en el nodo para seguir dividiéndolo
                             random_state=911)
clf = clf.fit(X_train, y_train)

# Calculamos la precisión del modelo de entrenamiento y de test
train_accuracy = clf.score(X_train, y_train)
test_accuracy = clf.score(X_test, y_test)
train_accuracy, test_accuracy
```
