Para esta parte utilizaremos el conjunto de datos USArrests, el cual describe los arrestos producidos en EEUU entre los distintos estados en funci√≥n de asesinatos, arrestos y violaciones producidas.

Dado que el algoritmo k-means es muy sensible al rango de los datos y que nuestro conjunto presenta un rango muy grande entre los tipos de culumna, estandarizamos nuestros datos con media = 0 y varianza = 1.

```python
#Escala

mat = df.values
scaler = StandardScaler()
mat = scaler.fit_transform(mat[:,1:4])
scaled_df = pd.DataFrame(mat, columns=['Murder','Assault','Rape'])
```

```python
def plot_clusters(X, labels=None, centers=None):
    """ Funci√≥n auxiliar para pintar los puntos de los clusters y, optativamente, sus centros.
    :param X: array de puntos de dos dimensiones (array de array de enteros)
    :param labels: cluster al que pertenece cada punto (array de enteros)
    :param centers: coordenadas de los centroides de los clusters (array de array de enteros)
    """
    
    colors = ['r','g','b','c','y','k','w','m',]
    fig = plt.figure(figsize=(5,5))
    
    # Los dos primeros par√°metros de la funci√≥n scatter son las coordenadas de los puntos,
    # el par√°metro 'c' indica la clase asignada de cada punto, y 'cmap' se usa para colorear 
    # las distintas clases
    plt.scatter(X[:,0], X[:,1], c=labels, cmap=ListedColormap(colors))
    
    # Pintar los centroides de los clusters
    if centers is not None:
        plt.scatter(centers[:,0], centers[:,1], marker="x", color='k', s=150, linewidths = 5, zorder=10)
        
    plt.show()
```

```python
#Clustering

K_MAX = 8
davies_boulding = np.zeros(K_MAX-2)
for k in range(2, K_MAX): 
    km = KMeans(init='random', n_clusters=k, random_state=rand)
    km.fit(scaled_df.values)
    plot_clusters(scaled_df.values, km.labels_, km.cluster_centers_)
    
    davies_boulding[k-2] = davies_bouldin_score(scaled_df.values, km.labels_) 
```

```python
#Valor de Davies Boulding para diferentes k

plt.plot(range(2, K_MAX), davies_boulding)
```

Dado el valor de Davies-Bouldin, observamos que 2 es el n√∫mero de clusters √≥ptimo dentro de un rango de de clusters razonable.

```python
clusters = 2 #Sacado del valor de Davies-Bouldin

km = KMeans(init='random', n_clusters=clusters, random_state=rand)
km.fit(scaled_df.values)
new_df = df.copy()
new_df['Cluster'] = km.labels_
```

```python
new_df.groupby('Cluster').describe().transpose()
```

```python
sns.set(style="ticks")
sns.pairplot(new_df, hue='Cluster', vars=['Murder','Assault','Rape'])
```

En general vemos que no hay ni una variable con la que se diferencien en su totalidad los clusters (todas tienen una region de confusi√≥n en el medio). En funci√≥n de las variables murder y assault no hay demasiada confusi√≥n entre los dos clusters, mientras que en funci√≥n de la variable rape se llega a confundir bastante m√°s, quedando los puntos entre 20 y 30 muy poco claros.
